```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv("covid_india_clean.csv")
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="Date", y="Total Confirmed", color="blue")
plt.title("Total COVID-19 Confirmed Cases in India (2020)")
plt.xlabel("Date")
plt.ylabel("Total Confirmed Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="Date", y="Total Recovered", label="Recovered", color="green")
sns.lineplot(data=df, x="Date", y="Total Deceased", label="Deceased", color="red")
plt.title("Recovered vs Deceased COVID-19 Cases in India (2020)")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['Active Cases'] = df['Total Confirmed'] - df['Total Recovered'] - df['Total Deceased']

plt.figure(figsize=(12,5))
sns.lineplot(data=df, x='Date', y='Active Cases', color='purple')
plt.title("Estimated Active COVID-19 Cases in India (2020)")
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insights:
- Recovery rate improved sharply from June onward.
- Deaths rose slowly and flattened in late 2020.
- Strong response indicated by the wide gap between recovered and deceased.
